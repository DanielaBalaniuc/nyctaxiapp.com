```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic time series data
np.random.seed(42)
date_range = pd.date_range(start='2023-01-01', periods=200, freq='D')
values = np.sin(np.linspace(0, 10 * np.pi, 200)) + np.random.normal(0, 0.2, 200)
values[::50] += 3  # inject periodic anomalies
df_ts = pd.DataFrame({'date': date_range, 'value': values})

# Compute rolling mean and std deviation
window = 10
df_ts['rolling_mean'] = df_ts['value'].rolling(window).mean()
df_ts['rolling_std'] = df_ts['value'].rolling(window).std()

# Detect outliers using z-score
threshold = 2.5
df_ts['z_score'] = (df_ts['value'] - df_ts['rolling_mean']) / df_ts['rolling_std']
df_ts['is_outlier'] = df_ts['z_score'].abs() > threshold

# Plotting
plt.figure(figsize=(14, 5))
plt.plot(df_ts['date'], df_ts['value'], label='Value', color='blue')
plt.scatter(df_ts[df_ts['is_outlier']]['date'], df_ts[df_ts['is_outlier']]['value'], color='red', label='Anomaly')
plt.title("Time Series with Detected Anomalies")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Time Series Anomaly Detection Project - NYC Taxi Data with LSTM Autoencoder

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, RepeatVector, TimeDistributed, Dense
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load NYC Taxi Data (hourly passenger count)
# Dataset: https://www.kaggle.com/datasets/ananthu017/nyc-taxi-trip-duration
# For demonstration, we'll simulate similar data
np.random.seed(0)
date_range = pd.date_range(start='2023-01-01', periods=1000, freq='H')
data = np.sin(np.linspace(0, 50, 1000)) + np.random.normal(0, 0.5, 1000)
data[400:405] += 5  # Inject anomalies
data[700:705] -= 5
df = pd.DataFrame({'timestamp': date_range, 'passenger_count': data})

# Step 2: Preprocessing
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)
scaler = MinMaxScaler()
df['scaled_count'] = scaler.fit_transform(df[['passenger_count']])

# Step 3: Create sequences for LSTM Autoencoder
def create_sequences(values, time_steps=24):
    sequences = []
    for i in range(len(values) - time_steps):
        sequences.append(values[i:i + time_steps])
    return np.array(sequences)

TIME_STEPS = 24
X = create_sequences(df['scaled_count'].values, TIME_STEPS)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Step 4: Build LSTM Autoencoder
model = Sequential([
    LSTM(64, activation='relu', input_shape=(TIME_STEPS, 1), return_sequences=True),
    LSTM(32, activation='relu', return_sequences=False),
    RepeatVector(TIME_STEPS),
    LSTM(32, activation='relu', return_sequences=True),
    LSTM(64, activation='relu', return_sequences=True),
    TimeDistributed(Dense(1))
])

model.compile(optimizer='adam', loss='mae')
model.summary()

# Step 5: Train the Model
history = model.fit(X, X, epochs=20, batch_size=32, validation_split=0.1, shuffle=False)

# Step 6: Reconstruction Error
X_pred = model.predict(X)
mse = np.mean(np.abs(X_pred - X), axis=(1, 2))
threshold = np.percentile(mse, 95)

# Step 7: Flag anomalies
df = df.iloc[TIME_STEPS:]
df['reconstruction_error'] = mse
df['anomaly'] = df['reconstruction_error'] > threshold

# Step 8: Visualization
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['passenger_count'], label='Passenger Count')
plt.scatter(df[df['anomaly']].index, df[df['anomaly']]['passenger_count'],
            color='red', label='Anomaly')
plt.title('NYC Taxi Passenger Count - Anomaly Detection with LSTM')
plt.legend()
plt.tight_layout()
plt.show()

# Step 9: Export
plt.savefig('nyc_taxi_lstm_anomaly.png')
df.to_csv('nyc_taxi_anomaly_output.csv')
```

```python
# NYC Taxi Real Data - LSTM Autoencoder Anomaly Detection

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, RepeatVector, TimeDistributed, Dense

# Step 1: Load NYC Taxi Data (adjust path)
df = pd.read_csv('yellow_tripdata_2016-01.csv', usecols=['tpep_pickup_datetime'])
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df.set_index('tpep_pickup_datetime', inplace=True)

# Step 2: Resample and count trips hourly
hourly_counts = df.resample('H').size().to_frame(name='passenger_count')
hourly_counts = hourly_counts.asfreq('H', fill_value=0)

# Step 3: Scale the data
scaler = MinMaxScaler()
hourly_counts['scaled_count'] = scaler.fit_transform(hourly_counts[['passenger_count']])

# Step 4: Create sequences for LSTM
TIME_STEPS = 24

def create_sequences(values, time_steps=TIME_STEPS):
    output = []
    for i in range(len(values) - time_steps):
        output.append(values[i:i + time_steps])
    return np.array(output)

X = create_sequences(hourly_counts['scaled_count'].values)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Step 5: Build LSTM Autoencoder
model = Sequential([
    LSTM(64, activation='relu', input_shape=(TIME_STEPS, 1), return_sequences=True),
    LSTM(32, activation='relu', return_sequences=False),
    RepeatVector(TIME_STEPS),
    LSTM(32, activation='relu', return_sequences=True),
    LSTM(64, activation='relu', return_sequences=True),
    TimeDistributed(Dense(1))
])

model.compile(optimizer='adam', loss='mae')
model.fit(X, X, epochs=10, batch_size=32, validation_split=0.1, shuffle=False)

# Step 6: Compute reconstruction error
X_pred = model.predict(X)
mse = np.mean(np.abs(X_pred - X), axis=(1, 2))
threshold = np.percentile(mse, 95)

# Step 7: Tag anomalies
hourly_counts = hourly_counts.iloc[TIME_STEPS:]
hourly_counts['reconstruction_error'] = mse
hourly_counts['anomaly'] = hourly_counts['reconstruction_error'] > threshold

# Step 8: Plot anomalies
plt.figure(figsize=(14, 6))
plt.plot(hourly_counts.index, hourly_counts['passenger_count'], label='Passenger Count')
plt.scatter(hourly_counts[hourly_counts['anomaly']].index,
            hourly_counts[hourly_counts['anomaly']]['passenger_count'],
            color='red', label='Anomaly')
plt.title('NYC Taxi Passenger Count - LSTM Anomaly Detection')
plt.legend()
plt.tight_layout()
plt.savefig('nyc_taxi_lstm_real_anomaly.png')
plt.show()

# Step 9: Export
hourly_counts.to_csv('nyc_taxi_real_anomalies.csv')
```

```python
# Streamlit App: NYC Taxi Anomaly Detector

import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.set_page_config(page_title="NYC Taxi Anomaly Detector", layout="wide")
st.title("🚕 NYC Taxi Passenger Count - Anomaly Detection")

# Load data
@st.cache_data
def load_data():
    # Assuming the first column (index 0) is the datetime column and there's no header
    df = pd.read_csv('nyc_taxi_real_anomalies.csv', header=None, usecols=[0, 1, 2, 3, 4])
    df.columns = ['tpep_pickup_datetime', 'passenger_count', 'scaled_count', 'reconstruction_error', 'anomaly']
    df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')
    print(f"Number of unparseable datetime values: {df['tpep_pickup_datetime'].isna().sum()}")
    # Drop rows where datetime is NaT before setting index and sorting
    df.dropna(subset=['tpep_pickup_datetime'], inplace=True)
    df.set_index('tpep_pickup_datetime', inplace=True)
    # Sort index to ensure monotonicity
    df.sort_index(inplace=True)

    # Convert relevant columns to numeric, coercing errors
    df['passenger_count'] = pd.to_numeric(df['passenger_count'], errors='coerce')
    df['scaled_count'] = pd.to_numeric(df['scaled_count'], errors='coerce')
    df['reconstruction_error'] = pd.to_numeric(df['reconstruction_error'], errors='coerce')

    return df

df = load_data()

# Sidebar controls
st.sidebar.header("Filters")
start_date = st.sidebar.date_input("Start Date", df.index.min().date())
end_date = st.sidebar.date_input("End Date", df.index.max().date())
threshold_slider = st.sidebar.slider("Anomaly Threshold (%)", 90, 99, 95)

# Filtered data
filtered_df = df.loc[str(start_date):str(end_date)]

# Apply new threshold
new_threshold = filtered_df['reconstruction_error'].quantile(threshold_slider / 100.0)
filtered_df['anomaly_custom'] = filtered_df['reconstruction_error'] > new_threshold

# Plot
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(filtered_df.index, filtered_df['passenger_count'], label='Passenger Count')
ax.scatter(filtered_df[filtered_df['anomaly_custom']].index,
           filtered_df[filtered_df['anomaly_custom']]['passenger_count'],
           color='red', label='Anomaly')
ax.set_title('Anomaly Detection with Adjustable Threshold')
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Passenger Count")
plt.xticks(rotation=45)
st.pyplot(fig)

# Show data table
with st.expander("📄 View Data Table"):
    st.dataframe(filtered_df[['passenger_count', 'reconstruction_error', 'anomaly_custom']])

# Download
st.download_button("Download Anomalies CSV", data=filtered_df.to_csv().encode('utf-8'), file_name="filtered_anomalies.csv", mime="text/csv")
```
